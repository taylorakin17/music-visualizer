```python
from note_pair import NotePair
from statistics import mean
```

```python
# Go through score and record the number of occurrences of each note
notes_dict_bassoon_1 = {}
note_pairs_b1 = []
note_pairs_b2 = []
num_accidentals = 0
prev_bsn_1 = None
prev_bsn_2 = None

for el in bsn_part.recurse().notes:
    if type(el)== music21.chord.Chord:
        sortedNotes = el.sortAscending()
        if prev_bsn_1 != None:
            note_pairs_b1.append(NotePair(prev_bsn_1.pitch.midi, sortedNotes[1].pitch.midi))
        if prev_bsn_2 != None:
            note_pairs_b2.append(NotePair(prev_bsn_2.pitch.midi, sortedNotes[0].pitch.midi))
        
        prev = sortedNotes[1]
        prev = sortedNotes[0]

    else:
        if prev_bsn_1 != None:
            note_pairs_b1.append(NotePair(prev_bsn_1.pitch.midi, el.pitch.midi))
            note_pairs_b2.append(NotePair(prev_bsn_2.pitch.midi, el.pitch.midi))
        prev_bsn_1 = el
        prev_bsn_2 = el

technical_difficulty_b1 = []
technical_difficulty_b2 = []

for pair in note_pairs_b1:
    technical_difficulty_b1.append(pair.difficulty)

for pair in note_pairs_b2:
    technical_difficulty_b2.append(pair.difficulty)

td_1 = "{:.2f}".format(mean(technical_difficulty_b1))
td_2 = "{:.2f}".format(mean(technical_difficulty_b2))

print("Technical difficulty for Part 1:", sum(technical_difficulty_b1))
print("Technical difficulty for Part 2:", sum(technical_difficulty_b2))
```
